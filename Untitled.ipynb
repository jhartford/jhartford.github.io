```python
class MLP(nn.Module):
    '''
    Modified from Karpathy's nanoGPT
    '''

    def __init__(self, n_embd, dropout=0., bias=True):
        super().__init__()
        self.c_fc    = nn.Linear(n_embd * 2, 4 * n_embd, bias=bias)
        self.gelu    = nn.GELU()
        self.c_proj  = nn.Linear(4 * n_embd, n_embd, bias=bias)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x_1, x_2):
        x = torch.cat([x_1, x_2], dim=-1)
        x = self.c_fc(x)
        x = self.gelu(x)
        x = self.c_proj(x)
        x = self.dropout(x)
        return x

class Pairwise(nn.Module):
    def __init__(self, base_linear: nn.Module, dropout=0.):
        super().__init__()
        self.n_classes, self.n_features = base_linear.weight.data.shape
        self.linear = nn.Linear(self.n_features, self.n_classes)
        self.linear.weight.data = base_linear.weight.data - base_linear.weight.data[0, :][None, :]
        self.linear.bias.data = base_linear.bias.data
        self.mlp = MLP(self.n_features, dropout=dropout)

    def forward(self, x):
        i,j = torch.triu_indices(self.n_classes, self.n_classes, offset=0)
        weights_i = self.linear.weight[i, :].T
        weights_j = self.linear.weight[j, :].T
        w_tilde = (weights_i + weights_j +  MLP(weights_i.detach(), weights_j.detach()))
        b_tilde = (l.bias[i] + l.bias[j])
        return x @ w_tilde + b_tilde
        
l  = nn.Linear(10, 20)
p = Pairwise(l)
x = torch.rand(7, 10)
p(x).shape
```
